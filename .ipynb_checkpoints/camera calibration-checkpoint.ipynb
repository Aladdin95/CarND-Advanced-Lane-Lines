```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import pickle
%matplotlib inline
```

```python
def view_imgs(imgs, cols=4):
    """ this function to show a list of imgs using subplot """
    rows = np.ceil(len(imgs)/cols)
    width = imgs[0].shape[1]
    height = imgs[0].shape[0]
    figsize = width/70, height/60
    plt.figure(figsize= figsize)
    for i, img in enumerate(imgs):
        camp = 'gray' if len(img.shape) < 3 else None
        plt.subplot(rows, cols, i+1)
        plt.imshow(img,cmap=camp)
    plt.tight_layout(pad=0, h_pad=0, w_pad=0)
    plt.show()
```

```python
calib_imgs = glob.glob('camera_cal/*.jpg')
```

```python
nx = 9
ny = 6

objp = np.zeros((nx*ny,3), np.float32)
objp[:,:2] = np.mgrid[0:nx, 0:ny].T.reshape(-1,2)

objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

out = []

for indx, img_name in enumerate(calib_imgs):
    img = cv2.imread(img_name)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    ret, corners = cv2.findChessboardCorners(gray, (nx,ny), None)
    
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw and display the corners
        cv2.drawChessboardCorners(img, (nx,ny), corners, ret)
        out.append(img)
        write_name = 'camera_cal/corners_found/' + str(indx) + '.jpg'
        cv2.imwrite(write_name, img)
        #plt.imshow(img)
        #cv2.imshow('img', img)
        #cv2.waitKey(5000)
    else:
        out.append(255*np.ones_like(img))
cv2.destroyAllWindows()
view_imgs(out)
```

```python
# Do camera calibration given object points and image points
img_size = img.shape[1::-1]
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_size,None,None)
```

```python
undistorted = []
for indx, img_name in enumerate(calib_imgs):
    img = cv2.imread(img_name)
    dst = cv2.undistort(img, mtx, dist, None, mtx)
    write_name = 'camera_cal/undistorted_images/' + str(indx) + '.jpg'
    cv2.imwrite(write_name,dst)
    undistorted.append(dst)
view_imgs(undistorted)
```

```python
dist_pickle = {}
dist_pickle["mtx"] = mtx
dist_pickle["dist"] = dist
pickle.dump( dist_pickle, open( "camera_cal/dist_pickle.p", "wb" ) )
```
